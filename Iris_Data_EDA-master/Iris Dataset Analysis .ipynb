```python
# Bee Swarm Plot
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
sns.swarmplot(x=data['label'],y=data['sepal length (cm)'])
```

```python
# Emperical Cumlative Distribution Function

import numpy as np
x = np.sort(data['petal length (cm)'])
y = np.arange(1, len(x)+1) / len(x)

plt.plot(x, y, marker='.', linestyle='none')
_ = plt.xlabel('petal length')
_ = plt.ylabel('ECDF')
plt.margins(0.02) # Keeps data off plot edges
plt.show()
```
